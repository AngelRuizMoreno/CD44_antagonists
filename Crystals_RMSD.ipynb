```python
import os
import matplotlib as mt
from matplotlib import pyplot as plt

import pandas as pd
```

## RMSD backbone

```python
path='../CD44_crystal_structures/RMSD/alpha_carbons/'
```

```python
m='../CD44_crystal_structures/RMSD/alpha_carbons/RMSD_Ca_1uuh_A_Crystal_human_vs_1uuh_B_Crystal_human.txt'
```

```python
table=pd.DataFrame()
with open (m,'r') as f:
    residue=[]
    rmsd=[]
    for line in f:
        line=line.split()
        if len(line)==2:
            residue.append (line[0])
            rmsd.append(line[1])
table['residue']=pd.Series(residue[:-1])
table['1uuh']=pd.Series(rmsd[:-1])
```

```python
for file in os.listdir (path):
    all_prots=pd.DataFrame()
    residue=[]
    rmsd=[]
    if '.txt' in file:
        with open (path+file,'r') as f:
            all_prots[file.split('_')[7]]=pd.Series()
            for line in f:
                line=line.split()
                if len(line)==2:
                    residue.append (line[0])
                    rmsd.append(line[1])
        all_prots['residue']=pd.Series(residue[:-1])
        all_prots[file.split('_')[7]]=pd.Series(rmsd[:-1])
        table=table.merge(all_prots,on='residue')
```

```python
table.to_csv('../CD44_crystal_structures/RMSD/alpha_carbons/all_rmsd.csv')
```

```python
for column in table.columns:
    x=[float(i) for i in table[column]]
    table[column]=x
```

```python
table.head()
```

```python
plt.rcParams['axes.linewidth'] = 1.5
plt.figure(figsize=(10,4))

for col in table.columns:
    
    if col in ['1uuh_x','4pz3_x','4pz3_y','4pz4_x','4pz4_y']:
        plt.plot ([i+22 for i in table['residue']],table[col],'k',label=col,linewidth=0.3)

    if col in ['2jcr','2jcq','4mrd']:
        plt.plot ([i+22 for i in table['residue']],table[col],'r',label=col,linewidth=0.5)
    
        
#plt.title ('RMSD Alpha Carbons',fontsize=14,fontweight='bold',family='sans-serif')
plt.text(150, 6.5, 'Alpha Carbons', fontsize=12,fontweight='bold')
#plt.xlabel ('Residue Î±C',fontsize=12,fontweight='bold')
plt.ylabel ('RMSD (Ã…)',fontsize=12,fontweight='bold')
plt.tick_params ('both',width=2,labelsize=12)
plt.legend(loc='upper left')
plt.grid (axis='y')
plt.tight_layout()
plt.plot()
```
